```python
from pyecharts.charts import Radar
from pyecharts import options as opts

v1 = [[4300, 10000, 28000, 35000, 50000, 19000]]
v2 = [[5000, 14000, 18000, 31000, 42000, 21000]]
c = (
    Radar(init_opts=opts.InitOpts(width="400px", height="400px"))
    .add_schema(
        schema=[
            opts.RadarIndicatorItem(name="Hp", max_=6500),
            opts.RadarIndicatorItem(name="Dps", max_=16000),
            opts.RadarIndicatorItem(name="Attack", max_=30000),
            opts.RadarIndicatorItem(name="Defence", max_=38000),
            opts.RadarIndicatorItem(name="Cost", max_=52000),
            opts.RadarIndicatorItem(name="Sp", max_=25000),
        ]
    )
    .add("阿能", v1)
    .add("小羊", v2)
)
c.render_notebook()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Sankey

nodes = [
    {"name": "Category1"},
    {"name": "Category2"},
    {"name": "Category3"},
    {"name": "Category4"},
    {"name": "Category5"},
    {"name": "Category6"},
]

links = [
    {"source": "Category1", "target": "Category2", "value": 5},
    {"source": "Category2", "target": "Category3", "value": 10},
    {"source": "Category3", "target": "Category4", "value": 15},
    {"source": "Category5", "target": "Category6", "value": 20}
]

c = (
    Sankey(init_opts=opts.InitOpts(width="600px", height="400px"))
    .add("sankey",
         nodes, links,
         linestyle_opt=opts.LineStyleOpts(opacity=0.2, curve=0.5, color="source"),
         label_opts=opts.LabelOpts(position="right"),
         )
    .set_global_opts(title_opts=opts.TitleOpts(title="Sankey_Sample"))
)
c.render_notebook()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Graph

nodes = [
    {"name": "A", "symbolSize": 50, "label": {"show": True}},
    {"name": "B", "symbolSize": 30, "label": {"show": True}},
    {"name": "C", "symbolSize": 40, "label": {"show": True}}
]

links = [
    {"source": "A", "target": "B", "tooltip": {"show": True, "formatter": "A B"}},
    {"source": "B", "target": "C", "tooltip": {"show": True, "formatter": "B C"}}
]

c = (
    Graph()
    .add("", nodes, links, repulsion=8000)
    .set_global_opts(title_opts=opts.TitleOpts(title="Title"))
)
c.render_notebook()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Graph

nodes = [
    {"name": "节点1", "symbolSize": 10},
    {"name": "节点2", "symbolSize": 20},
    {"name": "节点3", "symbolSize": 30},
    {"name": "节点4", "symbolSize": 40},
    {"name": "节点5", "symbolSize": 50},
    {"name": "节点6", "symbolSize": 60},
    {"name": "节点7", "symbolSize": 70},
    {"name": "节点8", "symbolSize": 80},
]
links = []
for i in nodes:
    for j in nodes:
        links.append(
                {"source": i.get("name"), "target": j.get("name")}
        )
        
c = (
    Graph()
    .add("", nodes, links, repulsion=8000)
    .set_global_opts(title_opts=opts.TitleOpts(title="Title"))
)
c.render_notebook()
```

```python
from pyecharts import options as opts
from pyecharts.charts import HeatMap
from pyecharts.faker import Faker
import random

values = [[i, j, random.randint(0, 50)] for i in range(24) for j in range(7)]
c = (
    HeatMap()
    .add_xaxis(Faker.clock)
    .add_yaxis(
        "series0",
        Faker.week,
        values,
        label_opts=opts.LabelOpts(is_show=True,position="inside"),
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="HeatMap-Label"),
        visualmap_opts=opts.VisualMapOpts(),
    )
)
c.render_notebook()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Geo
from pyecharts.faker import Faker

c = (
    Geo()
    .add_schema(maptype="china")
    .add("geo", [list(z) for z in zip(Faker.provinces, Faker.values())])
    .set_series_opts(label_opts=opts.LabelOpts(is_show=True))
    .set_global_opts(
        visualmap_opts=opts.VisualMapOpts(),
        title_opts=opts.TitleOpts(title="Geo-sample")
    )
)
c.render("./GeoMap.html")
```

```python
from pyecharts import options as opts
from pyecharts.charts import Geo
from pyecharts.globals import ChartType, SymbolType

c = (
    Geo()
    .add_schema(maptype="china")
    .add(
        "",
        [("广州", 50), ("北京", 100), ("杭州", 77), ("重庆", 88)],
        type_=ChartType.EFFECT_SCATTER,
        color="white"
    )
    .add(
        "geo",
        [("广州", "上海"), ("广州", "北京"), ("广州", "杭州"), ("广州", "重庆")],
        type_=ChartType.LINES,
        effect_opts=opts.EffectOpts(
            symbol=SymbolType.ARROW,
            symbol_size=6,
            color="red"
        ),
        linestyle_opts=opts.LineStyleOpts(curve=0.2)        
    )
    .set_series_opts(label_opts=opts.LabelOpts(is_show=True))
    .set_global_opts(title_opts=opts.TitleOpts(title="title"))
)
c.render("./Map-Arrow.html")
```
