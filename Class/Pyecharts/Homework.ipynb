根据上面数据做两种类型的玫瑰图，要求调整至少两次radius和center，加标题，图例垂直展示，调整位置不能遮挡图形


```python
from pyecharts import options as opts
from pyecharts.charts import Pie
from pyecharts.faker import Faker

Data=[60,74,85,78,92,81]
Labels=['化学','生物','物理','数学','英语','语文']
        
c = (
    Pie(init_opts=opts.InitOpts(width="1200px", height="600px"))
    .add(
        "",
        [list(z) for z in zip(Labels, Data)],
        radius=["25%", "75%"],
        center=["25%", "50%"],
    )
    .add(
        "",
        [list(z) for z in zip(Labels, Data)],
        radius=["45%", "75%"],
        center=["70%", "50%"],
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(title="成绩占比玫瑰图"),
        legend_opts=opts.LegendOpts(orient="vertical", pos_top="15%", pos_left="2%"),
    )
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}"))
)
c.render_notebook()
```

练习2：
    随意抓取一个论坛或者在线媒体、网页讨论的话题，统计出前20个关键词，基于Pyechart使用词云展示，要求给出数据爬取代码与截图。


```python
import requests
import re

URL = "https://weibo.com/ajax/side/hotSearch"

response = requests.get(url=URL).text

# 使用正则表达式提取中文汉字
chinese_pattern = re.compile(r'[\u4e00-\u9fa5]+')
chinese_texts = chinese_pattern.findall(response)

chinese_text = ''.join(chinese_texts)

print("提取到的中文文本：", chinese_text)

with open('./text.txt', 'w', encoding='utf-8') as f:
    f.write(chinese_text)
    f.close()
```

```python
import jieba
from collections import Counter
from pyecharts.charts import WordCloud

with open('./text.txt', 'r') as f:
    chinese_text = f.read()
    f.close()

segment = jieba.cut(chinese_text)

word_counts = Counter(segment)

wd = (
    WordCloud()
    .add("", word_counts.items(), word_size_range=[20, 100])
)

wd.render_notebook()
```
