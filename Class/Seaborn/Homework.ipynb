用以下‘tips’数据和catplot函数分别做散点图、箱线图：  
散点图要求分别对jitter，kind和调色板进行参数调整，出三幅图；  
箱线图要求分别kind、notch和背景以及调色板进行调整，出四幅图。

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')

tips = sns.load_dataset('tips')
print(tips.head())


sns.catplot(x="day", y="total_bill", jitter=False, data=tips)
plt.savefig('jitter_modify1.png')
plt.show()

sns.catplot(x="day", y="total_bill", kind="swarm", data=tips)
plt.savefig('kind_modify1.png')
plt.show()

sns.catplot(x="day", y="total_bill", palette="ch:r=-.2,d=.3_r", data=tips)
plt.savefig('palette_modify1.png')
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", kind="violin", data=tips)
plt.savefig('kind_modify2.png')
plt.show()

#缺口
sns.catplot(x="day", y="total_bill", kind="box", notch=True, data=tips)
plt.savefig('notch_modify2.png')
plt.show()

#背景
sns.set(style="darkgrid")
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
plt.savefig('background_modify2.png')
plt.show()

#调色板
sns.catplot(x="day", y="total_bill", kind="box", palette="ch:r=-.2,d=.3_r", data=tips)
plt.savefig('palette_modify2.png')
plt.show()
```
